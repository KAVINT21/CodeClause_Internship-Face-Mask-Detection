```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array

# Load the saved model
model = load_model("mask_detector.model")

# Load and preprocess the image
image_path = "with_mask_10.jpg"  # Replace with the actual path to your image
image = cv2.imread(image_path)


image = cv2.resize(image, (224, 224))
image = img_to_array(image)
image = np.expand_dims(image, axis=0)
image = np.array(image) / 255.0

# Make predictions on the image
(mask, withoutMask) = model.predict(image)[0]

# Determine the class label
label = "Mask" if mask > withoutMask else "No Mask"
confidence = max(mask, withoutMask) * 100

# Display the prediction result
print(f"Predicted: {label} with confidence: {confidence:.2f}%")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Display the loaded image
img = mpimg.imread(image_path)
plt.imshow(img)
plt.title(f"Predicted: {label} with confidence: {confidence:.2f}%")
plt.axis('off')
plt.show()
```
